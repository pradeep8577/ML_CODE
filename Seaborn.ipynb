```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset(load_dataset) and Plot(relplot)
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill", y = "tip", data = tips)
```

```python
# Use colouring using 'hue'
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", data = tips)
```

```python
# Plot using 'style' for different symbol
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", style = "smoker", data = tips)
```

```python
# Use 'style' for time(overlap data_variable)
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", style = "time", data = tips)
```

```python
# Using "catplot"
sns.catplot(x = "day", y = "total_bill", data = tips)
```

```python
# jitter false gives plot in one centered line
sns.catplot(x = "day", y = "total_bill", jitter = False, data = tips)
```

```python
# kind as swarm gives plot without overlapping
sns.catplot(x = "day", y = "total_bill", kind = "swarm", data = tips)
```

```python
# With hue and kind for sex
sns.catplot(x = "day", y = "total_bill", hue = "sex", kind = "swarm", data = tips)
```

```python
# With hue and kind for time
sns.catplot(x = "total_bill", y = "day", hue = "time", kind = "swarm", data = tips)
```

```python
# Using 'box' as kind
sns.catplot(x = "day", y = "total_bill", kind = "box", data = tips)
```

```python
# With hue as smoker
sns.catplot(x = "day", y = "total_bill", hue = "smoker", kind = "box", data = tips)
```

```python
# Diamonds dataset with kind as boxen (Box plot)
diamonds = sns.load_dataset("diamonds")
sns.catplot(x = "color", y = "price", kind = "boxen", data = diamonds.sort_values("color"))
```

```python
# titanic dataset
titanic = sns.load_dataset("titanic")
sns.catplot(x = "sex", y = "survived", hue = "class", kind = "bar", data = titanic)
```

```python
# using palette
sns.catplot(x = "deck", kind = "count", palette = "ch:.25", data = titanic)
```

```python
# Using pastel as palette and edgecolor .6
sns.catplot(y = "deck", hue = "class", kind = "count", palette = "pastel", edgecolor = ".6", data = titanic)
```

```python
# hue is class and kind is point 
sns.catplot(x = "sex", y = "survived", hue = "class", kind = "point", data = titanic)
```
